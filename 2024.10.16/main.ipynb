```python
def dell(x):
    M = set()
    for y in range(1,round(x ** 0.5) + 1):
        if x % y == 0:
            M |= {y, x // y}
    summ = sum(M)
    if summ > 10:
        if str(summ)[-2] == '7':
            return summ

for x in range(500_000,500_115):
    if dell(x):
        print(x, dell(x))
```

```python
graph = {
    1: {4: 30, 5: 3, 7: 5},
    2: {4: 21, 6: 13},
    3: {5: 39, 6: 53, 7: 2},
    4: {1: 30, 2: 21},
    5: {1: 3, 3: 39, 6: 8},
    6: {2: 13, 3: 53, 5: 8},
    7: {1: 5, 3: 2},
}
```

```python
for v, vs in graph.items():
    if len(vs) == 3 and all(len(graph[vv]) == 3 for vv in vs):
        print(v, '= G')
    if len(vs) == 2 and all(len(graph[vv]) == 3 for vv in vs):
        print(v, '= E')
    if len(vs) == 3 and 5 in vs and sum(len(graph[vv]) == 2 for vv in vs) == 2:
        print(v, '= C')
```

```python
from collections import Counter

data = [[int(x) for x in line.strip().split(";")] for line in open("9.csv").readlines()]
counter = 0
for line in data:
    count = Counter(line)
    if (
        len(set(line)) == 4
        and sum(x == 3 for x in count.values()) == 1
        and sum(c * x for x, c in count.items() if c > 1)**2
        > sum(x for x, c in count.items() if c == 1)**2
    ):
        counter += 1
print(counter)
# 273
```

```python
def F(x):
    s = ''
    while x > 0:
        s = str(x % 7) + s
        x //= 7
    return s
for x in range(1,2031):
    s = F(7 ** 170 + 7 ** 100 - x)
    if s.count('0') == 71:
        print(x)
# 2029
```

```python
inP = lambda x: 15 <= x <= 40
inQ = lambda x: 21 <= x <= 63
def F(x,A1,A2):
    inA = lambda x: A1 <= x <= A2
    return inP(x) <= ((inQ(x) and not(inA(x))) <= (not inP(x)))
minn = 999999999
for A1 in range(200):
    for A2 in range(A1 + 1, 200):
        if all(F(x,A1,A2) for x in range(200)):
            minn = min(minn, A2 - A1)
print(minn)
# 19
```
