```python
from itertools import product
from collections import Counter

def F(s):
    ms = s
    while '3' in ms:
        ms = ms.replace('342', '4123', 1)
        ms = ms.replace('34', '413', 1)
        ms = ms.replace('32', '13', 1)
        ms = ms.replace('33', '424', 1)
    return sum(map(int, ms)), s

print(max([F(f'3{"".join(x)}3') for x in product('42', repeat=20) if Counter(x)['2'] == 5], key=lambda x: x[0]))
print(max([F(f'3{"".join(x)}3') for x in product('42', repeat=20) if Counter(x)['2'] == 7], key=lambda x: x[0]))
print(max([F(f'3{"".join(x)}3') for x in product('42', repeat=20) if Counter(x)['2'] == 9], key=lambda x: x[0]))
```

```python
def F(x):
    while x % 2 == 0:
        x //= 2
    a = [1, x]
    for y in range(3,round(x ** 0.5) + 1, 2):
        if x % y == 0:
            a.append(y)
            z = x // y
            if z % 2 == 1:
                a.append(z)
        if len(a) > 5: break
    return a
            
```

```python
from tqdm import tqdm
res = []
for i in tqdm(range(45_000_000, 50_000_000 + 1)):
    if len(F(i)) == 5:
        res.append(i)
print(res)
```

```python
def F(x):
    while x % 2 == 0:
        x //= 2
    p = x**0.25
    if p == int(p):
        return [None] * 5
    return []
```

```python
res = []
for i in range(45_000_000, 50_000_000 + 1):
    if len(F(i)) == 5:
        res.append(i)
print(res)
```
