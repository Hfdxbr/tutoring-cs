```python
def F(x,y,z,w):
    return not(x <= w) or (y <= z) or not(y)
print('x y z w')
for x in range(2):
    for y in range(2):
        for z in range(2):
            for w in range(2):
                if F(x,y,z,w) == 0:
                    print(y,x, w, z)
# yxwz
```

```python
for n in range(1,1000):
    r = bin(n)[2:]
    if r.count('1') % 2 == 0:
        r += '0'
        r = '10' + r[2:]
    else:
        r += '1'
        r = '11' + r[2:]
    r2 = int(r,2)
    if r2 > 50:
        print(n)
        break
# 19
```

```python
from itertools import product
counter = 0
arr = product(range(8), repeat=5)
for x in arr:
    if x[0] % 2 == 0 and x[-1] != 2 and x[-1] != 6 and str(x).count('7') <= 2:
        counter += 1
print(counter)
```
