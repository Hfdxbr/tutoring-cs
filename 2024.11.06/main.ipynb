```python
arr = [list(map(float, line.split())) for line in open('27B.txt').readlines()]
```

```python
def dist(p1, p2):
    return abs(p1[0] - p2[0]) + abs(p1[1] - p2[1])
```

```python
# A
# clusters = [[(x, y) for x, y in arr if x > 0], [(x, y) for x, y in arr if x < 0]]
# B
clusters = [[], [], []]
for x, y in arr:
    if x > -2.5:
        clusters[0].append((x,y))
    elif y > -2:
        clusters[1].append((x,y))
    else:
        clusters[2].append((x,y))
```

```python
centroids = []
for cluster in clusters:
    centroid = None
    min_dist = 10**10
    for point in cluster:
        new_dist = sum(dist(point, other_point) for other_point in cluster)
        if new_dist < min_dist:
            min_dist = new_dist
            centroid = point
    centroids.append(centroid)
Px = sum(x for x, y in centroids) / len(centroids)
Py = sum(y for x, y in centroids) / len(centroids)
print(int(Px * 10_000), int(Py * 10_000))
```
