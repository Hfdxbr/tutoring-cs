```python
from functools import lru_cache


@lru_cache(maxsize=None)
def play(k, step):
    if step > 10:
        return 666
    if k == 42:
        return step - 1
    if k < 42:
        possible = [play(k + i, step + 1) for i in (1, 3, 7)]
    else:
        possible = [play(k - i, step + 1) for i in (1, 3, 7)]
    res = min(possible, key=lambda x: x if (x - step) % 2 == 0 else 100000 - x)
    return res


win_step = 4
for S in range(1, 100):
    if play(S, 1) == win_step:
        print(S)
```
