```python
import re
from itertools import permutations, product


def F(x, y, z, w):
    return not (y) and (x == (w <= z))


l = "xyzw"
arr = []
for x, y, z, w in product(range(2), repeat=4):
    arr.append([x, y, z, w, F(x, y, z, w)])

patterns = [
    r"101\d1",
    r"\d1\d\d1",
    r"\d0100",
]

for p1, p2, p3, p4 in permutations(range(4)):
    counter = 0
    for pattern in patterns:
        for *xs, f in arr:
            if re.match(pattern, f"{xs[p1]}{xs[p2]}{xs[p3]}{xs[p4]}{int(f)}"):
                counter += 1
                break
    if counter == 3:
        print(f"{l[p1]}{l[p2]}{l[p3]}{l[p4]}")
```

```python
def to8(x):
    s = ''
    while x > 0:
        s = str(x % 8) + s
        x //= 8
    return int(s)

def from8(x):
    return sum((8 ** i) * int(k) for i, k in enumerate(reversed(str(x))))
    
for i in range(from8(1000), from8(9999)):
    i = str(i)
    r_min = min(int(i[0]) + int(i[-1]), int(i[1]) + int(i[2]))
    r_max = max(int(i[0]) + int(i[-1]), int(i[1]) + int(i[2]))
    r_res = int(str(r_min) + str(r_max))
    if r_res == 317:
        print(i)
# 5872
```

```python
from collections import defaultdict
with open('9-194.csv') as f:
    data = [[int(x.strip('"')) for x in line.strip().split(',')] for line in f.readlines()]
counter = 0
number_counter = defaultdict(int)
for line in data:
    for i, x in enumerate(line):
        number_counter[(i, x)] += 1
for line in data:
    repeated = sum(line) - sum(set(line))
    indices = [i for i, x in enumerate(line) if x == repeated]
    if len(set(line)) == 4 and sum(line) < repeated * 4 and any(number_counter[(i, repeated)] > 70 for i in indices):
        counter += 1
print(counter)
```

```python
def is_prime(x):
    for i in range(2, int(x ** 0.5) + 1):
        if x % i == 0:
            return False
    return True

def proga(s):
    while '00' not in s:
        s = s.replace('02', '101', 1)
        s = s.replace('11', '2', 1)
        s = s.replace('12', '21', 1)
        s = s.replace('010', '00', 1)
    return sum(map(int, s))

for n in range(125, 200):
    s = '0' + '1' * n + '2' * n + '0'
    if is_prime(proga(s)):
        print(n)
        break
```
