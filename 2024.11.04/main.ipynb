```python
s = open('24.txt').read()
s = s.replace('-', '*')
s = s.split('*') # ['213', '', '123124', '0012545', ...]
maxlen = 0
curlen = 0
for x in s:
    if x == '':
        maxlen = max(maxlen, curlen)
        curlen = 0
    elif x[0] == '0' and len(x) != 1:
        # 125*643534*0
        maxlen = max(maxlen, curlen + 2)
        curlen = len(x.lstrip('0'))
    else:
        curlen += len(x) + (curlen > 0)
maxlen = max(maxlen, curlen)    
print(maxlen)
```

```python
f=open('24.txt')
s=f.read().strip()
s=s.replace('-', '*')
a=s.split('*')

st=''
kmax=0

for i in range(len(a)):
    if (a[i]!='') and  ((a[i][0]!='0') or (a[i]=='0')):
        # Формируем цепочку-кандидата
        if st!='':
            st = st + '*' + a[i]
        else:
            st = st + a[i]
        if st.find('*')!=-1:
            kmax=max(kmax, len(st))
    else:
        if a[i]!='' and a[i][0]=='0':
            if a[i].count('0')!=len(a[i]):
                a[i]=a[i].lstrip('0')
            else:
                a[i]='0'
            st=a[i]
        else:
            st = ''

print(kmax)
```

```python
import re
pattern = r'(?:0|[1-9][0-9]*)(?:[+*](?:0|[1-9][0-9]*))*'
s = open('24_17685.txt').read()
subs = re.findall(pattern, s)
kmax = 0
def find_max(evals):
    for i in range(len(evals)):
        for j in range(len(evals), i + 1, -1):
            if sum(evals[i:j]) == 0:
                return i, j
    return 0, 0
for ss in subs:
    evals = list(eval(x) for x in ss.split('+'))
    i, j = find_max(evals)
    kmax = max(kmax, len('+'.join(ss.split('+')[i:j])))
print(kmax)
# 169
```

```python
def dell(x):
    M = []
    for y in range(2,round(x ** 0.5) + 1):
        if x % y == 0:
            M.append(y)
            M.append(x // y)
    return max(M) + min(M) if M else 0

res = []
for i in range(800_000, 1_800_000):
    if dell(i) % 10 == 4:
        res.append([i, dell(i)])
        if len(res) == 5:
            break
print(*res, sep='\n')
```

```python
N, *arr = open('26.txt').readlines()
N = int(N)
arr = [list(map(int, line.split())) for line in arr]
```

```python
# ID M1 M2 M3 M4
def key(x):
    num_of_2 = x[1:].count(2)
    if num_of_2 > 0:
        return num_of_2, 0, x[0]
    else:
        return num_of_2, -sum(x[1:]) / 4, x[0]
arr.sort(key=key)
```
