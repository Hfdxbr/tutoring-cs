```python
def f(x,m):
    if x <= 33:
        return m % 2 == 0
    if m == 0:
        return 0
    h=[f(x-3, m-1),f(x-5, m-1),f(x // 3 + bool(x % 3), m-1)]
    return any(h) if m % 2 != 0 else all(h)
print("19",len([i for i in range(34, 10000) if f(i,2)]))
print("20",[i for i in range(34, 10000) if f(i,3) and not f(i,1)])
print("21",min([i for i in range(34, 10000) if not f(i,2) and f(i,4)]))
```

```python
def F(x, y, counter={30: 0, 60: 0}):
    if x > y:
        return 0
    if x == y and sum(counter.values()) == 1:
        return 1
    c = counter.copy()
    if x in c:
        c[x] += 1
    return F(x + 1, y, c) + F(x * 2, y, c) + F(x * 3, y, c)


print(F(10, 70))
```

```python
def f(n, t, d):
    if n > t or n == d:
        return 0
    if n == t:
        return 1
    return f(n + 1, t, d) + f(2 * n, t, d) + f(3 * n, t, d)


print(f(10, 30, 0) * f(30, 70, 60) + f(10, 60, 30) * f(60, 70, 0))
```

```python
def F(x, m):
    if x <= 19:
        return m % 2 == 0
    if m == 0:
        return 0
    h = [F(x - 1, m - 1)]
    h.append(F(x // 3, m - 1) if x % 3 == 0 else F(x - 2, m - 1))
    h.append(F(x // 5, m - 1) if x % 5 == 0 else F(x - 3, m - 1))
    return any(h) if m % 2 != 0 else all(h)


print([s for s in range(20, 1000) if F(s, 2)])
print([s for s in range(20, 1000) if F(s, 3) and not F(s, 1)])
print([s for s in range(20, 1000) if F(s, 4) and not F(s, 2)])
```

```python
def F(x,y,s = ''):
    if x > y:
        return 0
    if x == y and s != 'C':
        return 1
    return F(x + 2, y, 'A') + F(x + 5, y, 'B') + F(x ** 2, y, 'C')
print(F(4,36))
```
