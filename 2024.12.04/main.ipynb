```python
def F(x,y):
    s = ''
    while x > 0:
        s = str(x % y) + s
        x //= y
    return s

for n in range(1,1000):
    r = F(n,4)
    if n % 4 == 0:
        r = '2' + r + '03'
    else:
        r += F((n % 4) * 5, 4)
    if int(r,4) <= 567:
        print(n)
# 34
        
```

```python
from itertools import product
from collections import defaultdict
res = defaultdict(int)
arr = product(range(2), repeat = 20)
for x in arr:
    res[sum(x)] += 1
print(res)
```

```python
from string import ascii_uppercase

symbols = "".join(map(str, range(10))) + ascii_uppercase


def F(x, base):
    return sum(base**i * symbols.index(k) for i, k in enumerate(reversed(x)))


for x in symbols[:25]:
    if (res := F(f"A4{x}7F2", 25) + F(f"N{x}G5{x}H", 25) + F(f"74{x}M26", 25)) % 24 == 0:
        print(x, res // 24)
```

```python
def F(x,A1,A2):
    return (not((48 <= x <= 94) or (83 <= x <= 100))) <= (not(A1 <= x <= A2))
maxx = -9999999999999
for A1 in range(500):
    for A2 in range(A1 + 1, 500):
        if all(F(x,A1,A2) for x in range(500)):
            maxx = max(maxx, A2 - A1)
print(maxx)
```

```python

data = {}
def F(n):
    if n not in data:
        if n < 100:
            data[n] = n * 2
        if n > 99:
            coef = 0.5 if n % 2 == 0 else 2
            data[n] = coef * F(n - 1)
    return data[n]
for n in range(1,16385):
    F(n)
print((1000 * F(16384)) // F(7777))
```

```python
data = [int(x) for x in open("17.txt")]
minn = min(data) % 10
counter = 0
maxx = -999999999999999
for i in range(2, len(data)):
    m = data[i - 2 : i + 1]
    if sum(x < 0 for x in m) > sum(x > 0 for x in m) and sum(m) % 10 == minn:
        counter += 1
        maxx = max(maxx, sum(abs(x) for x in m))
print(counter, maxx)
```
