```python
data = [int(x) for x in open('17.txt')]
maxx = max(data)
maxx_final = -99999999999
counter = 0
for i in range(2, len(data)):
    pair = data[i - 2:i + 1]
    if sum(str(x).count('0') == 0 for x in pair) > 1 and sum(pair) < maxx / 2:
        counter += 1
        maxx_final = max(maxx_final, sum(pair))
print(counter, maxx_final)
```

```python
import re


def splitter(a):
    r = [[a[0]]]
    for v in a[1:]:
        if v == r[-1][-1] + 1:
            r[-1].append(v)
        else:
            r.append([v])
    return r

s = open("24.txt").read()
s = [len(x) for x in re.split("[AE]", s)]
s = splitter(s)
maxx = 0
for i, l in enumerate(s):
    curr = sum(l) + len(l) + 1
    if i > 0 and s[i - 1]:
        curr += s[i - 1][-1]
    if i < len(s) - 1 and s[i + 1]:
        curr += s[i + 1][0]
    maxx = max(maxx, curr)
print(maxx)
```

```python
from fnmatch import *
def F(x):
    M = []
    for y in range(1, int(x ** 0.5) + 1):
        if x % y == 0:
            M.append(y)
            M.append(x // y)
    S = [z for z in M if fnmatch(str(z), '2*3?')]
    return min(S) if S else None
for x in range(500_000, 510_000):
    if F(x):
        print(x, F(x))
        
# 500016 20834
# 500018 233
# 500020 230
# 500038 238
# 500040 20835
```
