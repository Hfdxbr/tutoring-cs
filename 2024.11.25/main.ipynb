```python
# НЕВЕРНОЕ УСЛОВИЕ ЧИСЕЛ 7
from math import prod
from collections import Counter
data = [[int(x) for x in line.strip().split(',')] for line in open('9.csv').readlines()]
counter = 0
for line in data:
    c = Counter(line)
    prd = prod(k**v for k, v in c.items() if v > 1)
    sm = sum(k for k, v in c.items() if v == 1)
    # if len(set(line)) < 6 and sum(x for x in line if line.count(x) == 1) * 3 <= prod(x for x in line if line.count(x) > 1):
    if len(set(line)) < 7 and sm * 3 <= prd:
        counter += 1

print(counter)
```

```python
def num(x,y):
    return x % 10 == y % 10

def F(x,A):
    return ((not num(x,5)) and num(x,4)) <= (x > A - 11)
for A in range(1,5000):
    if all(F(x,A) for x in range(1,5000)):
        print(A)
```

```python
data = {}
def F(n):
    if n not in data:
        if n > 3000:
            data[n] = n
        else:
            data[n] = (2 * n + 4) * F(n + 2)
    return data[n]
print(sum(int(x) for x in str(F(20) // F(28))))
```

```python
def cond(arr):
    if any(len(str(abs(x))) == 5 for x in arr) and prod(arr) % minn == 0:
        return prod(arr)


data = [int(x) for x in open("17.txt")]
minn = min([x for x in data if abs(x) % 10 == 7])
maxx = -9999999999
counter = 0
for i in range(2, len(data)):
    if prd := cond(data[i - 2 : i + 1]):
        counter += 1
        maxx = max(maxx, prd)
print(counter, maxx)
```
